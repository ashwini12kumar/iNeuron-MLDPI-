```python
import numpy as np
import pandas as pd
# import matplotlib.pyplot as plt
# from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

# Data Pre-Processing

```python
iris = datasets.load_iris()
iris.keys()
```

```python
df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
# iris['feature_names']
df.head()
```

```python
# Dimensionality reduction
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
df = pca.fit_transform(df)
df = pd.DataFrame(df)
df.head()
```

```python
df['species'] = pd.DataFrame(iris['target'])
df.head()
```

# Plotting

```python
#making interactive 
%matplotlib notebook  
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D   # importing Axes3D from mplot3d module of mpl_toolkits

fig = plt.figure(figsize=(10,6))   # Creating a new figure
ax = fig.add_subplot(111, projection='3d')   # Adding axes

ax.scatter(df[0],df[1],df[2], c=df['species'])
```
